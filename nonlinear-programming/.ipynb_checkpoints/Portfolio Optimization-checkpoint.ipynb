```python
import yfinance as yf
import numpy as np

# List of stocks in the portfolio
stocks = ['AAPL', 'AMZN', 'MSFT', 'GOOG']

def stock_info(stocks):
    # Download daily price data for each of the stocks in the portfolio
    data = yf.download(stocks, start='2010-01-01')['Adj Close']

    data.sort_index(inplace=True)

    # Convert daily stock prices into daily returns
    returns = data.pct_change()

    # Calculate mean daily return and covariance of daily returns
    mean_daily_returns = returns.mean()
    cov_matrix = returns.cov()

    
    return data, returns, mean_daily_returns, cov_matrix

# List of stocks in the portfolio
stocks = ['AAPL', 'AMZN', 'MSFT', 'GOOG']
data, returns, mean_daily_returns, cov_matrix = stock_info(stocks)
```

```python

# Set array holding portfolio weights of each stock
weights = np.asarray([0.5, 0.2, 0.2, 0.1])

def expected_return_volatility(mean_daily_returns, cov_matrix, weights):
    # Calculate annualized portfolio return
    portfolio_return = round(np.sum(mean_daily_returns * weights) * 252, 2)
    # Calculate annualized portfolio volatility
    portfolio_std_dev = round(np.sqrt(np.dot(weights.T, np.dot(cov_matrix, weights))) * np.sqrt(252), 2)
    print('Portfolio expected annualized return is {} and volatility is {}'.format(portfolio_return, portfolio_std_dev))
expected_return_volatility(mean_daily_returns, cov_matrix, weights)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Assuming mean_daily_returns and cov_matrix have been defined as per your previous code...

# Set number of runs of random portfolio weights
num_portfolios = 25000

# Set up array to hold results
results = np.zeros((3, num_portfolios))

for i in range(num_portfolios):
    # Select random weights for portfolio holdings
    weights = np.random.random(4)
    # Rebalance weights to sum to 1
    weights /= np.sum(weights)
    
    # Calculate portfolio return and volatility
    portfolio_return = np.sum(mean_daily_returns * weights) * 252
    portfolio_std_dev = np.sqrt(np.dot(weights.T, np.dot(cov_matrix, weights))) * np.sqrt(252)
    
    # Store results in results array
    results[0,i] = portfolio_return
    results[1,i] = portfolio_std_dev
    # Store Sharpe Ratio (return / volatility) - risk free rate element excluded for simplicity
    results[2,i] = results[0,i] / results[1,i]

# Convert results array to Pandas DataFrame
results_frame = pd.DataFrame(results.T, columns=['ret', 'stdev', 'sharpe'])

# Create scatter plot coloured by Sharpe Ratio
plt.figure(figsize=(12, 6))
plt.scatter(results_frame.stdev, results_frame.ret, c=results_frame.sharpe, cmap='RdYlBu')
plt.colorbar().set_label('Sharpe Ratio')
plt.xlabel('Portfolio Volatility (Standard Deviation)')
plt.ylabel('Portfolio Expected Annual Return')
plt.title('Portfolio Expected Return vs. Volatility Colored by Sharpe Ratio')
plt.grid(True)
plt.tight_layout()
plt.show()
# Instead of plt.show(), save the current figure object.
fig = plt.gcf()

```

```python
from gurobipy import *

# Define the Markowitz optimization function
def Markowitz(r, print_output = True):
    # Initialize a new model using Gurobi's API
    m = Model()
    # Suppress optimization output
    m.setParam('OutputFlag', 0)
    # Define variables for each stock's weight in the portfolio, ranging between 0 and 1
    x = m.addVars(stocks, name="x")

    # Add a constraint that the sum of the stock weights should be equal to 1
    # (i.e., the total allocation should use up 100% of the portfolio)
    m.addConstr(sum([x[i] for i in stocks]) == 1)
    
    # Calculate the expected portfolio return based on the stock weights and their respective mean daily returns
    obj_return = sum([mean_daily_returns[i]*x[i] for i in stocks])
    
    # Calculate the portfolio risk as the weighted combination of stock covariances
    obj_risk = sum([cov_matrix[i][j]*x[i]*x[j] for i in stocks for j in stocks])
    
    
    m.addConstr(obj_risk<= r)
    # Set the objective function to maximize the portfolio's return minus a penalty term for risk.
    # The penalty term is controlled by the parameter R (a measure of risk aversion).
    m.setObjective(obj_return, GRB.MAXIMIZE)
    
    # Optimize the model
    m.optimize()
    


    # Calculate the annualized risk of the optimal portfolio based on stock weights and covariances
    risk = np.sqrt(sum([cov_matrix[i][j]*x[i].x*x[j].x for i in stocks for j in stocks]))*np.sqrt(252)
    
    # Calculate the annualized expected return of the optimal portfolio
    expected_return = sum([mean_daily_returns[i]*x[i].x for i in stocks])*252

    if print_output:
        # Print the optimal stock weights
        for i in stocks:
            print(i + ' ' + str(x[i].x))
        # Print the annualized expected return and risk
        print("Expected Return: " + str(expected_return*252))
        print("Risk:" + str(risk*np.sqrt(252)))
    
    # Return the risk and expected return as a list
    return ([risk, expected_return], {i : x[i].x for i in stocks})
```

# Risk Range

```python
# Define the Markowitz optimization function
def Markowitz_risk():
    # Initialize a new model using Gurobi's API
    m = Model()
    # Suppress optimization output
    m.setParam('OutputFlag', 0)
    
    # Define variables for each stock's weight in the portfolio, ranging between 0 and 1
    x = m.addVars(stocks, name="x")

    # Add a constraint that the sum of the stock weights should be equal to 1
    # (i.e., the total allocation should use up 100% of the portfolio)
    m.addConstr(sum([x[i] for i in stocks]) == 1)
    
    # Calculate the expected portfolio return based on the stock weights and their respective mean daily returns
    obj_return = sum([mean_daily_returns[i]*x[i] for i in stocks])
    
    # Calculate the portfolio risk as the weighted combination of stock covariances
    obj_risk = sum([cov_matrix[i][j]*x[i]*x[j] for i in stocks for j in stocks])
    
    # Set the objective function to maximize the portfolio's return minus a penalty term for risk.
    # The penalty term is controlled by the parameter R (a measure of risk aversion).
    m.setObjective(obj_return, GRB.MAXIMIZE)
    
    # Optimize the model
    m.optimize()
    
    max_return_risk = sum([cov_matrix[i][j]*x[i].x*x[j].x for i in stocks for j in stocks])
    
    m.setObjective(obj_risk, GRB.MINIMIZE)
    m.update()
    
    # Optimize the model
    m.optimize()
    
    min_risk = m.obj_val
    return [min_risk, max_return_risk]

risk_range = Markowitz_risk()
print(risk_range)
```

```python
R = np.linspace(risk_range[0], risk_range[1], 30).tolist()

marks = []
for r in R:
    mark, solution = Markowitz(r,print_output = False)
    marks.append(mark)
```

```python
for mark in marks:
    plt.scatter([mark[0]], [mark[1]], color = 'g')
    # plt.ylim([0.017,0.02])
    # plt.xlim([0.05,0.1])
plt.show()
```
