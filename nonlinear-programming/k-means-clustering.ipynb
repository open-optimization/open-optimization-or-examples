# K-means Clustering
We show how to run k-means clustering using sklearn. 
We when then show how to model this problem as a mixed integer nonlinear program and solve it exactly.

The k-means algorithm is very effective!

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import pairwise_distances_argmin_min

# Generate synthetic data
np.random.seed(0)
points_A = np.random.randn(10, 2) + [1, 1]
points_B = np.random.randn(10, 2) + [5, 5]
points_C = np.random.randn(10, 2) + [8, 1]
points_D = np.random.randn(10, 2) + [3, 7]
points_E = np.random.randn(10, 2) + [7, 8]

data = np.vstack([points_A, points_B, points_C, points_D, points_E])
```

```python
# Apply k-means clustering
kmeans = KMeans(n_clusters=5)
kmeans.fit(data)
centroids = kmeans.cluster_centers_
labels = kmeans.labels_

colors = ["r.", "b.", "g.", "y.", "m."]

# Calculate radii for each cluster
radii = []
for i, center in enumerate(centroids):
    cluster_points = data[labels == i]
    distances = np.linalg.norm(cluster_points - center, axis=1)
    radii.append(np.max(distances))

# Plot
fig, ax = plt.subplots()
for i in range(len(data)):
    ax.plot(data[i][0], data[i][1], colors[labels[i]], markersize=10)

for i, center in enumerate(centroids):
    ax.scatter(center[0], center[1], s=50, c='black', marker='x')
    circle = plt.Circle((center[0], center[1]), radii[i], color=colors[i][0], fill=False)
    ax.add_artist(circle)

ax.set_aspect('equal', 'box')
plt.tight_layout()

# Export to TikZ
import tikzplotlib

tikzplotlib.save("test.tex")
```

# Integer Programming Implementation
The resulting integer program is a convex mixed-binary problem.  But this problem has many Big-M constraints, so the lower bounds are not very good.

```python
import gurobipy as gp
from gurobipy import GRB
import numpy as np

# Synthetic data as before...
# [data preparation code here]

n = 50
k = 5

# Create a model
m = gp.Model("kmeans_integer_linearized")

# Add decision variables
x = m.addVars(n, k, vtype=GRB.BINARY, name="x")
c = m.addVars(k, 2, lb=-GRB.INFINITY, name="c")
d = m.addVars(n, k, lb=0, name="d")

# Big M value - you may need to adjust this based on your dataset
M = 1000  

# Add objective
m.setObjective(gp.quicksum(gp.quicksum( d[i,j] for j in range(k)) for i in range(n)), GRB.MINIMIZE)

# Add constraints
for i in range(n):
    m.addConstr(gp.quicksum(x[i,j] for j in range(k)) == 1)

for i in range(n):
    for j in range(k):
        m.addConstr(d[i,j] >= np.sum((data[i,l] - c[j,l])*(data[i,l] - c[j,l]) for l in range(2)) - M*(1 - x[i,j]))

#for j in range(k):
    #for l in range(2):  # 2 dimensions
        #m.addConstr( gp.quicksum(x[i,j] for i in range(n))*c[j,l] == gp.quicksum(x[i,j] * data[i,l] for i in range(n)))

# Optimize model
m.optimize()

# Compute k-means objective value
kmeans_obj = sum(np.min(np.sum((data - centroid)**2, axis=1)) for centroid in centroids)
ip_obj = m.objVal

print(f"Objective value for k-means: {kmeans_obj}")
print(f"Objective value for IP: {ip_obj}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Assuming you have already solved the IP model (m) and have your data available

# Extract the centroids from the integer program solution
optimal_centroids = np.array([[c[j,0].X, c[j,1].X] for j in range(k)])

# Assign each data point to its nearest centroid based on IP solution
clusters = {i: [] for i in range(k)}
for i in range(n):
    for j in range(k):
        if x[i, j].X > 0.5:  # assuming a small tolerance
            clusters[j].append(data[i])
            break

# Plotting
fig, ax = plt.subplots()

# Plot data points
for j, cluster_points in clusters.items():
    cluster_array = np.array(cluster_points)
    ax.scatter(cluster_array[:, 0], cluster_array[:, 1], label=f"Cluster {j+1}")

# Plot centroids
ax.scatter(optimal_centroids[:, 0], optimal_centroids[:, 1], c='black', marker='x', s=100, label="Centroids")

ax.set_title("Clusters from Integer Program")
ax.legend()
plt.show()
```
