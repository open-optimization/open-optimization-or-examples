```python
import matplotlib.pyplot as plt
import numpy as np
import tikzplotlib
```

```python
X = np.arange(-60,28,0.05)

def f(x):
    return -2.5*x + -50
def g(x):
    return (-1/20)*x - 5/12
def h(x):
    return (15/4)*x -259/4

U = 35
c1 = 1.5
c2 = 1
c3 = 5
def myrange(X):
    return [min(X), max(X)]

plt.figure(figsize=(7, 4))
for d in [0,8,16,32,64]:
    X1 = [x for x in X if (g(x) <= f(x)+c1*d and h(x) <= f(x)+c1*d and f(x)+c1*d <= U)]
    
    X2 = [x for x in X if (f(x) <= g(x)+c2*d and h(x) <= g(x)+c2*d and g(x)+c2*d <= U)]
    
    X3 = [x for x in X if (f(x) <= h(x)+c3*d and g(x) <= h(x)+c3*d and h(x)+c3*d <= U)]
    
    if X1:
        X1 = myrange(X1)
        plt.plot(X1, [f(x)+c1*d for x in X1], c = 'b')
    if X2:
        X2 = myrange(X2)
        plt.plot(X2, [g(x)+c2*d for x in X2], 'g')
    if X3:
        X3 = myrange(X3)
        plt.plot(X3, [h(x)+c3*d for x in X3], 'purple')
    #plt.plot(x, -(1/12)*x + 5/12)
    #plt.plot(x, (15/4)*x -259/4)
plt.axis('off')

z = 2*np.array([-4.1,7.9]) - [-10,12]

plt.scatter([-10],[12])
plt.annotate('p', xy=(-10, 14)   )

plt.scatter([z[0]],[z[1]])
plt.annotate('2q-p', xy=(0, 5.5))

plt.scatter([-4.1],[7.9])
plt.annotate('q', xy=(-6,6))
tikzplotlib.save("reflecting.tex")
plt.savefig('reflecting_pdf.pdf',bbox_inches='tight',pad_inches=0)
plt.show()
```
