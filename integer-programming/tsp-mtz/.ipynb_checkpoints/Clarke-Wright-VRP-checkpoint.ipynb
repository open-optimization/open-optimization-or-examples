```python
import math
import matplotlib.pyplot as plt

def euclidean_distance(p1, p2):
    return math.sqrt((p1[0] - p2[0])**2 + (p1[1] - p2[1])**2)

def calculate_savings(customers, depot):
    savings = []
    for i in range(len(customers)):
        for j in range(i+1, len(customers)):
            s_ij = euclidean_distance(depot, customers[i]) + euclidean_distance(depot, customers[j]) - euclidean_distance(customers[i], customers[j])
            savings.append((s_ij, i, j))
    return sorted(savings, reverse=True, key=lambda x: x[0])

def clarke_wright_savings(customers, depot):
    routes = [[i] for i in range(len(customers))]
    savings = calculate_savings(customers, depot)
    
    for saving in savings:
        _, i, j = saving
        route_i = next((route for route in routes if i in route), None)
        route_j = next((route for route in routes if j in route), None)
        
        if route_i != route_j:
            route_i.extend(route_j)
            routes.remove(route_j)
            
    routes = [[0] + route + [0] for route in routes]
    return routes

def plot_routes(routes, customers, depot):
    plt.figure(figsize=(10, 7))
    for route in routes:
        x_coords = [depot[0]]
        y_coords = [depot[1]]
        for customer_idx in route[1:-1]:  # Excluding depot at start and end
            x_coords.append(customers[customer_idx][0])
            y_coords.append(customers[customer_idx][1])
        x_coords.append(depot[0])  # Add depot coordinates at the end
        y_coords.append(depot[1])
        plt.plot(x_coords, y_coords, marker="o")

    plt.scatter(depot[0], depot[1], c='red', s=100, label="Depot")
    plt.title("Routes using Clarke-Wright Savings Algorithm")
    plt.xlabel("X coordinate")
    plt.ylabel("Y coordinate")
    plt.legend()
    plt.grid(True)
    plt.show()

# Example
customers = [(2,3), (5,7), (8,6), (6,4)]
depot = (0, 0)
routes = clarke_wright_savings(customers, depot)
print("Routes:", routes)
plot_routes(routes, customers, depot)
```

```python
import math
import matplotlib.pyplot as plt

def euclidean_distance(p1, p2):
    return math.sqrt((p1[0] - p2[0])**2 + (p1[1] - p2[1])**2)

def calculate_savings_for_depot(customers, depot):
    savings = []
    for i in range(len(customers)):
        for j in range(i+1, len(customers)):
            s_ij = euclidean_distance(depot, customers[i]) + euclidean_distance(depot, customers[j]) - euclidean_distance(customers[i], customers[j])
            savings.append((s_ij, i, j))
    return sorted(savings, reverse=True, key=lambda x: x[0])

def clarke_wright_savings_for_depot(customers, depot, demands, capacity, assigned_customers):
    routes = [[i] for i in range(len(customers)) if i not in assigned_customers]
    savings = calculate_savings_for_depot(customers, depot)
    
    for saving in savings:
        _, i, j = saving
        if i in assigned_customers or j in assigned_customers:
            continue
        route_i = next((route for route in routes if i in route), None)
        route_j = next((route for route in routes if j in route), None)
        
        if route_i != route_j:
            if sum(demands[cust] for cust in route_i) + sum(demands[cust] for cust in route_j) <= capacity:
                route_i.extend(route_j)
                routes.remove(route_j)
                assigned_customers.add(i)
                assigned_customers.add(j)
            
    routes = [[depot] + route + [depot] for route in routes]
    return routes

def plot_routes(routes, customers, depots):
    plt.figure(figsize=(10, 7))
    for route in routes:
        x_coords = [route[0][0]]
        y_coords = [route[0][1]]
        for customer_idx in route[1:-1]:  # Excluding depot at start and end
            x_coords.append(customers[customer_idx][0])
            y_coords.append(customers[customer_idx][1])
        x_coords.append(route[-1][0])  # Add depot coordinates at the end
        y_coords.append(route[-1][1])
        plt.plot(x_coords, y_coords, marker="o")
    
    for depot in depots:
        plt.scatter(depot[0], depot[1], c='red', s=100, label=f"Depot {depots.index(depot)}")
    
    plt.title("Routes using Clarke-Wright Savings Algorithm with Multiple Depots")
    plt.xlabel("X coordinate")
    plt.ylabel("Y coordinate")
    plt.legend()
    plt.grid(True)
    plt.show()

# Example
customers = [(2,3), (5,7), (8,6), (6,4), (3,8), (7,9)]
demands = [3, 4, 2, 5, 3, 2]  # Demands of customers
vehicle_capacity = 5  # Vehicle capacity constraint
depots = [(0, 0), (10, 10)]
all_routes = []
assigned_customers = set()

for depot in depots:
    routes_for_depot = clarke_wright_savings_for_depot(customers, depot, demands, vehicle_capacity, assigned_customers)
    all_routes.extend(routes_for_depot)

print("Routes:", all_routes)
plot_routes(all_routes, customers, depots)
```

```python
# ... [previous code remains unchanged]

def clarke_wright_savings_interleaved(customers, depots, demands, capacity):
    all_routes = []
    assigned_customers = set()
    
    savings_for_depots = {depot: calculate_savings_for_depot(customers, depot) for depot in depots}
    while any(savings_for_depots.values()):
        for depot in depots:
            if not savings_for_depots[depot]:
                continue
                
            _, i, j = savings_for_depots[depot].pop(0)
            
            if i in assigned_customers or j in assigned_customers:
                continue
            
            route_i = next((route for route in all_routes if i in route and depot in route), None)
            route_j = next((route for route in all_routes if j in route and depot in route), None)
            
            if route_i is None:
                route_i = [depot, i, depot]
                all_routes.append(route_i)
            
            if route_j is None:
                route_j = [depot, j, depot]
                all_routes.append(route_j)
            
            if route_i != route_j:
                if sum(demands[cust] for cust in route_i[1:-1]) + sum(demands[cust] for cust in route_j[1:-1]) <= capacity:
                    route_i.insert(-1, j)
                    all_routes.remove(route_j)
                    assigned_customers.add(i)
                    assigned_customers.add(j)
    
    return all_routes

# Example
customers = [(2,3), (5,7), (8,6), (6,4), (3,8), (7,9)]
demands = [3, 4, 2, 5, 3, 2]  # Demands of customers
vehicle_capacity = 8  # Vehicle capacity constraint
depots = [(0, 0), (10, 10)]

all_routes = clarke_wright_savings_interleaved(customers, depots, demands, vehicle_capacity)

print("Routes:", all_routes)
plot_routes(all_routes, customers, depots)
```

```python
def clarke_wright_savings_interleaved(customers, depots, demands, capacity):
    all_routes = []
    
    savings_for_depots = {depot: calculate_savings_for_depot(customers, depot) for depot in depots}
    while any(savings_for_depots.values()):
        for depot in depots:
            if not savings_for_depots[depot]:
                continue
                
            _, i, j = savings_for_depots[depot].pop(0)
            
            existing_route_i = next((route for route in all_routes if i in route and route[0] == depot), None)
            existing_route_j = next((route for route in all_routes if j in route and route[0] == depot), None)
            
            if not existing_route_i:
                route_i = [depot, i, depot]
                all_routes.append(route_i)
            else:
                route_i = existing_route_i
            
            if not existing_route_j:
                route_j = [depot, j, depot]
                all_routes.append(route_j)
            else:
                route_j = existing_route_j
            
            if route_i != route_j:
                if sum(demands[cust] for cust in route_i[1:-1]) + sum(demands[cust] for cust in route_j[1:-1]) <= capacity:
                    route_i.insert(-1, j)
                    all_routes.remove(route_j)
            
            # Pruning process
            for customer_idx, routes in [(i, [route_i, route_j]), (j, [route_j, route_i])]:
                other_route = next((route for route in all_routes if customer_idx in route and route[0] != depot), None)
                if other_route:
                    dist_current = euclidean_distance(depot, customers[customer_idx])
                    dist_other = euclidean_distance(other_route[0], customers[customer_idx])
                    if dist_current < dist_other:
                        other_route.remove(customer_idx)
                    else:
                        if customer_idx in routes[0]:
                            routes[0].remove(customer_idx)
                        if customer_idx in routes[1]:
                            routes[1].remove(customer_idx)

    # Filter out routes with only depot (no customers)
    all_routes = [route for route in all_routes if len(route) > 2]
    return all_routes
# Example
customers = [(2,3), (5,7), (8,6), (6,4), (3,8), (7,9)]
demands = [3, 4, 2, 5, 3, 2]  # Demands of customers
vehicle_capacity = 8  # Vehicle capacity constraint
depots = [(0, 0), (10, 10)]

all_routes = clarke_wright_savings_interleaved(customers, depots, demands, vehicle_capacity)

print("Routes:", all_routes)
plot_routes(all_routes, customers, depots)
```

```python
import random

# ... [previous code remains unchanged]

def clarke_wright_with_randomization(customers, depots, demands, capacity, iterations=100):
    best_routes = []
    best_distance = float('inf')
    
    for _ in range(iterations):
        # Shuffle the order of customers and depots
        random.shuffle(customers)
        random.shuffle(depots)
        
        # Calculate routes using the interleaved method
        current_routes = clarke_wright_savings_interleaved(customers, depots, demands, capacity)
        
        # Calculate total distance of current solution
        current_distance = sum(sum(euclidean_distance(customers[route[i]], customers[route[i+1]]) for i in range(1, len(route)-2)) for route in current_routes)
        
        # Update best solution if current solution is better
        if current_distance < best_distance:
            best_distance = current_distance
            best_routes = current_routes
            
    return best_routes

# Example
customers = [(2,3), (5,7), (8,6), (6,4), (3,8), (7,9)]
demands = [3, 4, 2, 5, 3, 2]  # Demands of customers
vehicle_capacity = 9  # Vehicle capacity constraint
depots = [(0, 0), (10, 10)]

all_routes = clarke_wright_with_randomization(customers, depots, demands, vehicle_capacity, iterations=100)

print("Best Routes:", all_routes)
plot_routes(all_routes, customers, depots)
```

```python
def two_opt_vrp(routes, customers, demands, capacity, max_iterations=1000):
    def get_customer_or_depot(route_entry):
        if isinstance(route_entry, tuple):  # it's a depot
            return route_entry
        return customers[route_entry]

    iteration = 0
    improvement = True
    while improvement and iteration < max_iterations:
        improvement = False
        iteration += 1
        for index_r1, r1 in enumerate(routes):
            for index_r2, r2 in enumerate(routes):
                i = 1
                while i < len(r1) - 1:
                    j = 1
                    while j < len(r2) - 1:
                        if r1 == r2 and (abs(i-j) <= 1):
                            j += 1
                            continue
                        if (index_r1 != index_r2) and \
                           (sum(demands[cust] for cust in r1[1:i] if isinstance(cust, int)) + sum(demands[cust] for cust in r2[j+1:-1] if isinstance(cust, int)) > capacity or \
                            sum(demands[cust] for cust in r2[1:j] if isinstance(cust, int)) + sum(demands[cust] for cust in r1[i+1:-1] if isinstance(cust, int)) > capacity):
                            j += 1
                            continue
                        old_distance = euclidean_distance(get_customer_or_depot(r1[i-1]), get_customer_or_depot(r1[i])) + \
                                       euclidean_distance(get_customer_or_depot(r2[j-1]), get_customer_or_depot(r2[j]))
                        new_distance = euclidean_distance(get_customer_or_depot(r1[i-1]), get_customer_or_depot(r2[j])) + \
                                       euclidean_distance(get_customer_or_depot(r2[j-1]), get_customer_or_depot(r1[i]))
                        
                        if new_distance < old_distance:
                            if r1 == r2:
                                r1[i:j+1] = reversed(r1[i:j+1])
                            else:
                                r1_tail = r1[i+1:]
                                r1[i+1:] = r2[j+1:-1]
                                r2[j+1:-1] = r1_tail
                            improvement = True
                            break
                        j += 1
                    if improvement:
                        break
                    i += 1

    routes = [route for route in routes if len(route) > 2]
    return routes




# ... [Rest of the code remains unchanged]

all_routes = clarke_wright_with_randomization(customers, depots, demands, vehicle_capacity, iterations=100)
optimized_routes = two_opt_vrp(all_routes, customers, demands, vehicle_capacity)

print("Optimized Routes after 2-opt:", optimized_routes)
plot_routes(optimized_routes, customers, depots)
```
