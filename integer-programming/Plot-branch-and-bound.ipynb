```python
import numpy as np
import matplotlib.pyplot as plt

# Create a grid of points
x1 = np.linspace(0, 10, 400)
x2 = np.linspace(0, 10, 400)
X1, X2 = np.meshgrid(x1, x2)

# Define the inequalities
ineq1 = 2*X1 + 2*X2 <= 13
ineq2 = -8*X1 + 10*X2 <= 41
ineq3 = 9*X1 + 5*X2 <= 45

# Plotting
plt.figure(figsize=(10, 10))

# Plot each inequality
plt.contourf(X1, X2, ineq1, colors='blue', alpha=0.3, levels=[0, 1])
plt.contourf(X1, X2, ineq2, colors='red', alpha=0.3, levels=[0, 1])
plt.contourf(X1, X2, ineq3, colors='green', alpha=0.3, levels=[0, 1])

# Highlight integer points within the feasible region
for i in range(11):
    for j in range(11):
        if ineq1[i, j] and ineq2[i, j] and ineq3[i, j]:
            plt.scatter(i, j, color='black', s=50)

# Setting labels and title
plt.xlabel("$x_1$")
plt.ylabel("$x_2$")
plt.title("Feasible Region")
plt.xlim(0, 10)
plt.ylim(0, 10)
plt.grid(True)
plt.show()
```
